```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
circle=plt.Circle((-.005,.005),1 ,color='r')
fig = plt.gcf() 
fig.gca().add_artist(circle) 
plt.xlabel('AREA')
plt.ylabel('AREA')

n = 100000
x = np.random.uniform(-1,1,n)
y = np.random.uniform(-1,1,n)
lista = []

for j in range(n):
    lista.append(x)
    
    
plt.plot(x,y,'o',markersize=1)
axes = plt.gca()
axes.set_xlim([-1,1])
axes.set_ylim([-1,1])
plt.show()
```

```python
def reg_circ(x,y):
    return x**2+y**2<=1

A = 4
Aprox = A*np.sum(reg_circ(x,y))/n
Aprox
```
